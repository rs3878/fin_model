```python
import pandas as pd
import numpy as np
import xlrd
```

```python
data = pd.read_excel("Assignment 3 - Data.xlsx")
data.to_csv("data.csv")
df = pd.read_csv("data.csv", header = None)
```

```python
# find the rows that corresponds to the name of the hurricanes
from xlrd import open_workbook
wb = open_workbook('Assignment 3 - Data.xlsx')
for s in wb.sheets():
    names = []
    for row in range(s.nrows):
        for col in range(s.ncols):
            if col == 4 and s.cell(row,col).value in (None,""):
                names.append(row)
```

```python
# create a dictionary that records the three cells that contain the basic info of each hurricane
d = {}
for s in wb.sheets():
    for row in range(s.nrows):
        if row in names:
            d[row] = {}
            d[row]['id'] = s.cell(row,0).value
            d[row]['name'] = s.cell(row,1).value
            d[row]['Number of best track entries'] = s.cell(row,2).value        
```

```python
# create three new columns that has the basic info and apply to every row of the according hurricane
first_,second_,third_ = [],[],[]
for date, row in df.T.iteritems():
    compare = np.array(names) - date
    try:
        inde = list(compare).index(list(filter(lambda x: x>0, compare))[0]) 
    except:
        inde = len(names)
    index = names[inde-1]
    first_.append(d[index]['id'])
    second_.append(d[index]['name'])
    third_.append(d[index]['Number of best track entries'])
df['ID'] = pd.Series(first_, index=df.index)
df['Name'] = pd.Series(second_, index=df.index)
df['Number of best track entries'] = pd.Series(third_, index=df.index)
```

```python
# drop unused rows and columns
df = df.drop(df.index[names])
df = df.drop(columns=[0,21])
# reindex
df = df.reindex()
# change column type
df[1] = df[1].astype('datetime64')
```

```python
# change column names
new_name = {1:'Date', 4:'Status of System', 5:'Latitude_Hemisphere', 6:'Longitude_Hemisphere', 
            7:'Maximum sustained wind', 8:'Minimum Pressure', 
            9:'34 kt wind radii maximum extent in northeastern quadrant',
            10:'34 kt wind radii maximum extent in southeastern quadrant',
            11:'34 kt wind radii maximum extent in southwestern quadrant',
            12:'34 kt wind radii maximum extent in northwestern quadrant',
            13:'50 kt wind radii maximum extent in northeastern quadrant',
            14:'50 kt wind radii maximum extent in southeastern quadrant',
            15:'50 kt wind radii maximum extent in southwestern quadrant',
            16:'50 kt wind radii maximum extent in northwestern quadrant',
            17:'64 kt wind radii maximum extent in northeastern quadrant',
            18:'64 kt wind radii maximum extent in southeastern quadrant',
            19:'64 kt wind radii maximum extent in southwestern quadrant',
            20:'64 kt wind radii maximum extent in northwestern quadrant'
           }
df.rename(columns=new_name, inplace=True)
```

```python
X = df.drop(['Date', 2, 3, 'ID', 'Name', 'Number of best track entries'], axis = 1)
X.replace('-999.0', np.NaN, inplace = True)
X.replace(-999.0, np.NaN, inplace = True)
X.dropna(axis = 0, inplace = True)
X.reset_index(drop = True, inplace = True)
X['Maximum sustained wind'] = X['Maximum sustained wind'].astype(int)
X['Minimum Pressure'] = X['Minimum Pressure'].astype(int)
X.head()
```

```python
y = X[['Latitude_Hemisphere', 'Longitude_Hemisphere']]
y['Latitude_Number'] = y['Latitude_Hemisphere'].apply(lambda x: float(str(x).strip()[:-1]))
y['Latitude_N/S'] = y['Latitude_Hemisphere'].apply(lambda x: str(x).strip()[-1])
y['Longitude_Number'] = y['Longitude_Hemisphere'].apply(lambda x: float(str(x).strip()[:-1]))
y['Longitude_W/E'] = y['Longitude_Hemisphere'].apply(lambda x: str(x).strip()[-1])
y.drop(['Latitude_Hemisphere', 'Longitude_Hemisphere'], axis = 1, inplace = True)
y.head()
```

```python
X.drop(['Latitude_Hemisphere', 'Longitude_Hemisphere'],axis = 1, inplace = True)
X.head()
```

```python
X_train = pd.get_dummies(X, drop_first=True)
X_train.head()
```

### Predict Latitude_Number

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestRegressor

y1 = y['Latitude_Number']
scores = cross_validate(RandomForestRegressor(), X_train, y1, cv=5, scoring=('r2', 'neg_mean_squared_error'), return_train_score=True)
print(-np.mean(scores['test_neg_mean_squared_error']))     
print(np.mean(scores['train_r2']))                    
```

### Predict Longitude_Number

```python
y2 = y['Longitude_Number']
scores = cross_validate(RandomForestRegressor(), X_train, y2, cv=5, scoring=('r2', 'neg_mean_squared_error'), return_train_score=True)
print(-np.mean(scores['test_neg_mean_squared_error']))     
print(np.mean(scores['train_r2'])) 
```

### Predict Longitude_W/E

```python
from sklearn.ensemble import RandomForestClassifier
y3 = y['Longitude_W/E']
scores = cross_validate(RandomForestClassifier(), X_train, y3, cv=5, scoring=('accuracy'), return_train_score=True)    
print(np.mean(scores['test_score']))
```
